```python
#Removed duplicates Existed 
df.drop_duplicates(inplace = True)
```

```python
#Dropped unnnesary column
df.drop(columns = "Not_Useful_Column", inplace = True)
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('...') # Cleaning and removing using .str.lstrip - leftstrip, and assigning only to collumn df['Last_Name']
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
```

```python
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=True) # Removing unnessary characters from phone no. , Function is Regex
```

```python
#another way 
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[-/]', '', regex=True)
```

```python
df['Phone_Number'].apply(lambda x : str(x)) #converting number to a string
```

```python
df['Phone_Number'] =df['Phone_Number'].astype(str) # better alternative
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #adding - in between
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--','') #replacing nan-- with blank
df['Phone_Number'] = df['Phone_Number'].replace('Na--','')
```

```python
df['Phone_Number'] #done with phone number
```

```python
df['Address'].str.split(',', n = 2, expand = True) #Splitted the address
```

```python
df[['Street Address','State', 'Zip_Code']] = df['Address'].str.split(',', n = 2, expand = True)
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Y', 'Yes')

df['Paying Customer']= df['Paying Customer'].replace('N', 'No')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Y','Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N','No')
```

```python
df = df.replace('N/a','') #replacing N/a with blanks on (entire table)
```

```python
df = df.fillna('') # Fills NaN
df
```

```python
df = df[df['Do_Not_Contact'].isin(['No',''])] #filtering
```

```python
df.drop(columns = 'Address', inplace = True)
```

```python
df.reset_index(drop = True, inplace = True)
```
